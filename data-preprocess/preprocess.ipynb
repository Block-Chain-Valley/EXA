```python
import pandas as pd
import numpy as np
import requests
import json
import csv
import ccxt
```

```python
price= pd.DataFrame(columns=['symbol','price'])
binance = ccxt.binance()
binancecoinm = ccxt.binancecoinm()
binanceusdm = ccxt.binanceusdm()

binancemarkets = binance.fetch_tickers()
binancecoinmmarkets = binancecoinm.fetch_tickers()
binanceusdmmarkets = binanceusdm.fetch_tickers()
```

```python
denomination =['USDT','BUSD','USDC']
for bmarkets in binancemarkets.values():
    symbols= bmarkets['symbol'].split('/')
    if(symbols[1] in denomination):
        add = pd.DataFrame(data = {'symbol':[symbols[0]],'price':[bmarkets['last']]})
        price = pd.concat([price, add], ignore_index=True)

for bcmarkets in binancecoinmmarkets.values():
    symbols= bcmarkets['symbol'].split('/')
    if(symbols[1] in denomination):
        add = pd.DataFrame(data = {'symbol':[symbols[0]],'price':[bcmarkets['last']]})
        price = pd.concat([price, add], ignore_index=True)
price.to_csv('./Price.csv')
price = price.drop_duplicates(subset=['symbol'])
price
```

```python
df = pd.read_csv('./filterWithLiquidityV2.csv')
```

```python
df2 = pd.read_csv('./BinanceThicker.csv')
```

```python
overlap = []
for thicker in df2['Binance thicker']:
    for token in df['baseToken']:
        if thicker in token:
            overlap.append(token)
```

```python
remove = []
for i in range(len(df['baseToken'])):
    if df['baseToken'][i] not in overlap:
        remove.append(i)
remove.reverse()    
for r in remove:
        df.drop(df.index[r],inplace =True)
```

```python
df=df.drop(['Unnamed: 0','Unnamed: 1','index','TRUE'],axis ='columns',inplace = False)
df= df.reset_index(drop=True)
```

```python
drop_pairAddress = ["0xf20d962a6c8f70c731bd838a3a388D7d48fA6e15"]
df = df[~df['pairAddress'].isin(drop_pairAddress)]
```

```python
df
```

```python
#price와의 차이가 10%일때 남기기
for i in range(len(df)):
    price_dex_ = df.iloc[i]['liquidity'].split(',',':')
    price_dex = float(price_dex_[1])
    price_cex = price[price['symbol'] == df.iloc[i]['baseToken']].values[0][1]
    if(abs((price_dex) - price_cex)/price_cex > 0.1):
        df = df.drop(i)
```
