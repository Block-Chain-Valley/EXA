```python
import pandas as pd
import numpy as np
import requests
import json
import csv
import ccxt
import json
```

```python
price= pd.DataFrame(columns=['symbol','price'])
binance = ccxt.binance()
binancecoinm = ccxt.binancecoinm()
binanceusdm = ccxt.binanceusdm()

binancemarkets = binance.fetch_tickers()
binancecoinmmarkets = binancecoinm.fetch_tickers()
binanceusdmmarkets = binanceusdm.fetch_tickers()

denomination =['USDT','BUSD']

for bmarkets in binancemarkets.values():
    symbols= bmarkets['symbol'].split('/')
    if(symbols[1] in denomination):
        add = pd.DataFrame(data = {'symbol':[symbols[0]],'price':[bmarkets['last']]})
        price = pd.concat([price, add], ignore_index=True)

for bcmarkets in binancecoinmmarkets.values():
    symbols= bcmarkets['symbol'].split('/')
    if(symbols[1] in denomination):
        add = pd.DataFrame(data = {'symbol':[symbols[0]],'price':[bcmarkets['last']]})
        price = pd.concat([price, add], ignore_index=True)
        
for bcmarkets in binanceusdmmarkets.values():
    symbols= bcmarkets['symbol'].split('/')
    if(symbols[1] in denomination):
        add = pd.DataFrame(data = {'symbol':[symbols[0]],'price':[bcmarkets['last']]})
        price = pd.concat([price, add], ignore_index=True)

price = price.drop_duplicates(subset=['symbol'])
price = price[price['price']!= 0]
price.to_csv('./Price.csv')
price.reset_index(drop=True)
```
