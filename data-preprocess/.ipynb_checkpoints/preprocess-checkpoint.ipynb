```python
import pandas as pd
import numpy as np
import requests
import json
import csv
```

```python
df = pd.read_csv('./filterWithLiquidityV2.csv')
df
```

```python
df2 = pd.read_csv('./BinanceThicker.csv')
df2
```

```python
overlap = []
for thicker in df2['Binance thicker']:
    for token in df['baseToken']:
        if thicker in token:
            overlap.append(token)
overlap
```

```python
remove = []
for i in range(len(df['baseToken'])):
    if df['baseToken'][i] not in overlap:
        remove.append(i)
remove.reverse()    
for r in remove:
        df.drop(df.index[r],inplace =True)
```

```python
df
```

```python
df.drop(['Unnamed: 0','Unnamed: 1','index'],axis ='columns')
drop_pairAddress = ["0xf20d962a6c8f70c731bd838a3a388D7d48fA6e15"]
for i in range(len(df['baseToken'])):
    if df.drop['pairAddress'][i] not in overlap:
        remove.append(i)
```

```python
df.to_csv('./Final.csv')
```
